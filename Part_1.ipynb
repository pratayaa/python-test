```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
def plotting(stocks, weeks = [4,16,28,40,52]):
    df1 = pd.DataFrame()
    df1['Close'] = stocks['Close'].resample('W').mean()
    for i in range(len(weeks)):
        x = df1['Close'].rolling(weeks[i]).mean()
        df1["moving average for " + str(weeks[i]) +  "weeks"] = x
        print('calculated moving averages : for {0} weeks: \n\n {1} '.format(weeks[i], df1['Close']))
    df1.plot(title = "moving averages for {} \n\n".format(stocks.name))
```

```python
plotting(TCS)
```

```python
plotting(INFY)
```

```python
plotting(NIFTY)
```

```python
def plotting_rolling_window(stocks, window = [10, 75]):
    df1 = pd.DataFrame()
    df1['Close'] = stocks['Close']
    for i in range(len(window)):
        x = df1['Close'].rolling(win[i]).mean()
        df1["moving average for" + str(window[i]) + "roll window"] = x
        print('calculated moving averages : for {0} weeks: \n\n{1}'.format(window[i], df1['Close']))
    df1.plot(title = "moving averages for {}  \n\n".format(stocks.name))
```

```python
def volume_shocks(stocks):
    stocks["volume_t+1"] = stocks.Volume.shift(1)  
    stocks["volume_shock"] = ((abs(stocks["volume_t+1"] - stocks["Volume"])/stocks["Volume"]*100)  > 10).astype(int)
    return stocks
```

```python
volume_shocks(TCS)
volume_shocks(INFY)
volume_shocks(NIFTY)
```

```python
def direction_of_shock(stocks):
    if stocks["volume_shock"] == 0:
        pass
    else:
        if(stocks["volume_t+1"] - stocks["Volume"]) < 0:
            return 0
        else:
            return 1
```

```python
def volume_shock_direction(stocks):
    stocks['VOLUME_SHOCK_DIR'] = 'Nan'
    stocks['VOLUME_SHOCK_DIR'] = stocks.apply(direction_of_shock, axis = 1)
    return stocks
```

```python
volume_shock_direction(TCS)
volume_shock_direction(INFY)
volume_shock_direction(NIFTY)
```

```python
def Price_shocks(stocks):
    stocks["Price_t+1"] = stocks.Close.shift(1)
    stocks["Price_shock"] = (abs((stocks["Price_t+1"] -stocks["Close"])/stocks["Close"]*100) > 2).astype(int)
    stocks["Price_Black_Swan"] = stocks["Price_shock"]
    return stocks
```

```python
Price_shocks(TCS)
Price_shocks(INFY)
Price_shocks(NIFTY)
```

```python
def direction_of_shock_price(stocks):
    if stocks["Price_shock"] == 0:
        pass
    else:
        if(stocks["Price_t+1"] - stocks["Close"]) < 0:
            return 0
        else:
            return 1
```

```python
def price_shock_direction(stocks):
    stocks['PRICE_SHOCK_DIR'] = 'Nan'
    stocks['PRICE_SHOCK_DIR'] = stocks.apply(direction_of_shock_price, axis=1)
    return stocks
```

```python
price_shock_direction(TCS)
price_shock_direction(INFY)
price_shock_direction(NIFTY)
```

```python
def pricing_shock_without_volume_shock(stocks):
    stocks["not_vol_shock"]  = (~(stocks["volume_shock"].astype(bool))).astype(int)
    stocks["price_shock_without_vol_shock"] = stocks["not_vol_shock"] & stocks["Price_shock"]
    return stocks
```

```python
pricing_shock_without_volume_shock(TCS)
pricing_shock_without_volume_shock(INFY)
pricing_shock_without_volume_shock(NIFTY)
```
